```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
an_img = cv.imread(r'D:\DATASCIENCE\DeepLearning-DS\Chapter 03\OpenCV\76370213.jpg')

plt.figure(figsize=(10,5))
plt.imshow(an_img)
```

```python
rgb_an_img = cv.cvtColor(an_img, cv.COLOR_BGR2RGB)

plt.figure(figsize=(10,15))
plt.imshow(rgb_an_img)
```

```python
an_img_copy = rgb_an_img.copy()

cv.rectangle(an_img_copy, (470, 400), (1100,610), (255,0,0), 3)
cv.putText(an_img_copy, 'Boat', (470, 400), cv.FONT_HERSHEY_PLAIN, 3, (255,0,0), 2)

cv.rectangle(an_img_copy, (170,280), (570,530), (0,0,0), 3)
cv.putText(an_img_copy, 'Hotel', (170, 280), cv.FONT_HERSHEY_PLAIN, 3, (0,0,0), 2)


cv.rectangle(an_img_copy, (200,500), (420, 640), (0,0, 255), 3)
cv.putText(an_img_copy, 'High Rider', (200,500), cv.FONT_HERSHEY_PLAIN, 2, (0,0,255), 2)

# Line annotation
pt1 = (400, 800)
pt2 = (800,800)

cv.line(an_img_copy, pt1, pt2, (255,255,0), 3, cv.LINE_AA)
cv.putText(an_img_copy, 'River', (400,800), cv.FONT_HERSHEY_PLAIN, 3, (255,255,0), 2)


# Ellipse annotation

center = (600, 150) 
ax = (170, 100)

cv.ellipse(an_img_copy, center, ax, 0, 0, 360, (0,255,255), 3)
cv.putText(an_img_copy, 'Sky', (500,150), cv.FONT_HERSHEY_PLAIN, 3, (0,255,255), 2)


plt.figure(figsize=(10,15))
plt.imshow(an_img_copy)
```
