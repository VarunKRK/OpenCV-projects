```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
qr_img = cv.imread(r'D:\DATASCIENCE\DeepLearning-DS\Chapter 03\OpenCV\qr-code.png')

plt.imshow(qr_img)
```

```python
decoder = cv.QRCodeDetector()

data, points, _ = decoder.detectAndDecode(qr_img)
 
if points is not None:
    print('Decoded data: ' + data)
    points = points[0]
    for i in range(len(points)):
        pt1 = [int(val) for val in points[i]]
        pt2 = [int(val) for val in points[(i + 1) % 4]]
        cv.line(qr_img, pt1, pt2, color=(0, 255,0), thickness=3)
 
cv.imshow('Detected QR code', qr_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
